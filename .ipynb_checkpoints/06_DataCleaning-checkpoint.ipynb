```python
df = pd.read_csv("05_flats.csv")
df.drop_duplicates(inplace=True)
df.drop([286], inplace=True)
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].str.strip()
        df[col] = df[col].str.lower()
df['bhk'] = df['property_name'].str.extract(r"(\d*)")
df['locality'] = df['property_name'].str.split(" in ").apply(lambda x : x[1]).str.strip()
df = df.drop(columns=['property_name'])
df['price1'] = df['price'].str.extract(r"(\d*)")
df['price_unit'] = df['price'].str.replace(r"[^a-zA-Z]", "", regex=True)
df = df[(df['price_unit'].str.contains('lac')) | (df['price_unit'].str.contains('crore'))]

df.shape
```
